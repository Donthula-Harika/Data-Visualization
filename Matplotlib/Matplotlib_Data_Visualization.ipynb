 Handling Missing Values

I will use the mean imputation method to handle missing values in the 'Calories' column. This method replaces missing values with the mean of the existing values in that column. This is a simple and effective approach when the missing data is assumed to be missing at random and the distribution of the data is not heavily skewed by outliers.

```python
# 4. Handle missing values in numeric columns

# Check for missing values in numeric columns
numeric_cols = df.select_dtypes(include=np.number).columns
print("\nMissing values in numeric columns before handling:")
print(df[numeric_cols].isnull().sum())

# Fill missing values in 'Calories' with the mean
mean_calories = df['Calories'].mean()
df['Calories'].fillna(mean_calories, inplace=True)

print("\nMissing values in numeric columns after handling:")
print(df[numeric_cols].isnull().sum())

print("\nFirst few rows after handling missing values:")
display(df.head())
```

**Line Plot**

**Scatter plot**
- Plot a scatter plot of Duration vs Calories, compute the Pearson correlation coefficient, and add a linear trend line to the scatter; save as scatter_duration_vs_calories.png.

```python
plt.style.use('ggplot')
y = df['Duration']
x = df['Calories']

plt.figure(figsize=(10, 6)) # Optional: Adjust figure size for better readability
plt.scatter(x, y, color="blue", marker='o')

# Compute Pearson correlation coefficient
correlation_coefficient = x.corr(y)
print(f"Pearson correlation coefficient between Duration and Calories: {correlation_coefficient:.2f}")

# Add a linear trend line
z = np.polyfit(x, y, 1)
p = np.poly1d(z)
plt.plot(x, p(x), color="red", linestyle='--') # Add the trend line

plt.title("Duration vs Calories with Linear Trend")
plt.xlabel("Duration")
plt.ylabel("Calories")
plt.grid(True)
plt.savefig("scatter_duration_vs_calories.png") # Save the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import pearsonr

x = df['Duration']
y = df['Calories']

# Compute Pearson correlation
corr, _ = pearsonr(x, y)
print(f"Pearson Correlation Coefficient: {corr:.3f}")

# Create scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='teal', alpha=0.7, edgecolor='black', label='Data points')

# Fit and plot linear trend line
m, b = np.polyfit(x, y, 1)
plt.plot(x, m*x + b, color='crimson', linewidth=2, label='Trend line')

# Labels and styling
plt.title(f"Duration vs Calories (r = {corr:.2f})")
plt.xlabel("Duration")
plt.ylabel("Calories")
plt.legend()
plt.grid(True)
plt.tight_layout()

# Save the figure
# plt.savefig("scatter_duration_vs_calories.png")
plt.show()
```

**Pie chart**
- Create a pie chart showing the percentage breakdown of Maxpulse into ranges (e.g., <=120, 121140, 141160, >160) and save as pie_maxpulse_ranges.png.


```python
# Create Maxpulse ranges
bins = [0, 120, 140, 160, df['Maxpulse'].max()]
labels = ['<=120', '121-140', '141-160', '>160']
df['Maxpulse_Range'] = pd.cut(df['Maxpulse'], bins=bins, labels=labels, right=True, include_lowest=True)

# Count occurrences in each range
maxpulse_counts = df['Maxpulse_Range'].value_counts()

# Create the pie chart

plt.pie(maxpulse_counts, labels=maxpulse_counts.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired(range(len(maxpulse_counts))))
plt.title("Maxpulse Range Distribution")
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Save the figure
plt.savefig("pie_maxpulse_ranges.png")
plt.show()
```
